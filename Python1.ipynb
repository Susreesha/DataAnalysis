```python
# Install required libraries
!pip install pandas
import pandas as pd
import numpy as np

# Create the dataframe again so it is in a known state.
df1 = pd.read_excel("/content/Patient footfall (1).xlsx")

def check_ratio():
    # Iterate through rows and update the new column
    for index, row in df1.iterrows():
        ratio = row['Total Patients in IPD'] / row['Population within 10km']

        if ratio > 0.3:
            df1.loc[index,'ratio1'] = "greater"
        else:
            df1.loc[index,'ratio1'] = "lesser"
#Call the function.
check_ratio()

#Display the results
df1
```
