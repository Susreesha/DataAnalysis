```python
# mean and median of BMI Column

mean_bmi = dataframe_heartattack['BMI'].mean()
median_bmi = dataframe_heartattack['BMI'].median()
```

```python
dataframe_heartattack['BMI'] = dataframe_heartattack['BMI'].fillna(mean_bmi)
```

```python
dataframe_heartattack
```

```python
mean_heartrate = dataframe_heartattack['Heart Rate'].mean()
median_heartrate = dataframe_heartattack['Heart Rate'].median()
```

```python
dataframe_heartattack['Heart Rate'] = dataframe_heartattack['Heart Rate'].fillna(median_heartrate)
```

```python
dataframe_heartattack
```

**Remove Outliers**

```python
# Remove Outliers based on Cholesterol column

Q1 = dataframe_heartattack['Heart Rate'].quantile(0.25)
Q3 = dataframe_heartattack['Heart Rate'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

dataframe_heartattack = dataframe_heartattack[(dataframe_heartattack['Heart Rate'] >= lower_bound) & (dataframe_heartattack['Heart Rate'] <= upper_bound)]
```

```python
dataframe_heartattack
```

**Remove Duplicates**

```python
# drop duplicate entries based on patient id

dataframe_heartattack = dataframe_heartattack.drop_duplicates(subset=['Patient ID'])
dataframe_heartattack
```

```python
# drop duplicate entries based on country

dataframe_heartattack = dataframe_heartattack.drop_duplicates(subset=['Country'])
dataframe_heartattack
```

```python
dataframe_heartattack = dataframe_heartattack.drop(columns ='Age')
dataframe_heartattack
```

```python
dataframe_heartattack = dataframe_heartattack.rename(columns = {'BMI':'Body Mass Index'})
dataframe_heartattack
```

```python
dataframe_heartattack = dataframe_heartattack.rename(columns={'Heart Rate': 'Heart_Rate'})
dataframe_heartattack
```

**FILTERING**

```python
# Filtering based on column name

listColumns = ['Patient ID', 'Heart_Rate']
dataframe_heartattack = dataframe_heartattack[listColumns]
dataframe_heartattack
```

```python
# filter age > 50

filtered_dataframe_heartattack = dataframe_heartattack[dataframe_heartattack['Age'] > 50]
filtered_dataframe_heartattack.head()
```

**NORMALIZATION AND STANDARDIZATION**

```python
# Normalize and Standardize BMI Column

from sklearn.preprocessing import MinMaxScaler, StandardScaler

#Normalize BMI
scaler = MinMaxScaler()
dataframe_heartattack['BMI_Normalized'] = scaler.fit_transform(dataframe_heartattack[['BMI']])

#Standardize BMI
scaler = StandardScaler()
dataframe_heartattack['BMI_Standardized'] = scaler.fit_transform(dataframe_heartattack[['BMI']])

dataframe_heartattack[['BMI', 'BMI_Normalized', 'BMI_Standardized']]
```

**MAP VALUES**

```python
dataframe_heartattack['Gender'].unique()
```

```python
# Replace values in Gender Column with Male by M and Female by F

# Create a mapping dictionary
gender_mapping = {'Male': 'M', 'Female': 'F'}

# Use the map function to replace values in the 'Gender' column
dataframe_heartattack['Gender'] = dataframe_heartattack['Gender'].map(gender_mapping)

# Display the Updated Dataframe
dataframe_heartattack.head(5)
```

```python
dataframe_heartattack['Heart Risk Score Category'].unique()
```

```python
# Replace values in Heart Risk Score Category

risk_mapping = {'Low Risk': 0, 'Average Risk' : 1, 'High Risk': 2}
dataframe_heartattack['Heart Risk Score Category'] = dataframe_heartattack['Heart Risk Score Category'].map(risk_mapping)
dataframe_heartattack
```

```python
# Replace values in Heart Risk Score Category

risk_mapping = {'Low Risk': 0}
dataframe_heartattack['Heart Risk Score Category'] = dataframe_heartattack['Heart Risk Score Category'].replace(risk_mapping)
dataframe_heartattack
```

```python
# Prompt: apply a function on a column
# Define a function to apply

def sq_root(x):
  if x >= 0:
    return x**0.5
  else:
    return 0

# Apply the function to the Age Column
dataframe_heartattack['Age_sqrt'] = dataframe_heartattack['Age'].apply(sq_root)

# Display the updated Dataframe
dataframe_heartattack
```

```python
# prompt : apply a function on bmi column

def bmi_group(x):
  if x < 18.5:
    return 'Underweight'
  elif x >= 18.5 and x < 24.9:
    return 'Normal Weight'
  else:
    return 'Obesity'

dataframe_heartattack['BMI_Group'] = dataframe_heartattack['BMI'].apply(bmi_group)
dataframe_heartattack
```

```python
# prompt : apply a function on bmi column

def bmi_group(x):
  if x < 18.5:
    return 'Underweight'
  elif x >= 18.5 and x < 24.9:
    return 'Normal Weight'
  else:
    return 'Obesity'

dataframe_heartattack['BMI_Group'] = dataframe_heartattack['BMI'].apply(bmi_group)
dataframe_heartattack

dictagg = {'BMI': 'count'}
df_bmi_agg = dataframe_heartattack.groupby(by = ['BMI_Group','Age']).agg(dictagg)
```

```python
# prompt : apply a function on 2 columns with ifelse

# define a function to aplly based on conditions from two columns
def apply_function(row):
  if row['Age'] > 50 and row['Cholesterol'] > 200:
    return row['Age'] * 2
  elif row['Age'] <= 50 and row['Cholesterol'] <= 200:
    return row['Age'] /2
  else:
    return 0

# Apply the function to create a new column 'New Column'
dataframe_heartattack['New Column'] = dataframe_heartattack.apply(apply_function, axis =1)

# Display the updated DataFrame
dataframe_heartattack
```
