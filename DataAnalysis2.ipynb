```python
# print unique columns in a dataframe column

#Syntax
#df['column'].unique()

dist_unique = dataframe_nfhs['District'].unique()
dist_unique
```

```python
# Filtering based on column values

#Syntax
#df[df['column'] == value]

df_div = dataframe_nfhs[dataframe_nfhs['Division'] == "Bangalore"]
df_div
```

```python
# Filtering based on multiple conditions

# Syntax
# df[(df['column1'] = value1) & (df['column2'] = value2)]

df_div = dataframe_nfhs[(dataframe_nfhs['Division'] == "Mysore") & (dataframe_nfhs['Population living in households with an improved drinkingwater source'] > 50) & (dataframe_nfhs['Children under 5 years who are stunted']>20)]
df_div.head(1)
```

```python
# Min, Max, Mean, Std, Median

# Syntax
# df['column1'].min()

print("Minimum is:", dataframe_nfhs['Women who are literate'].min())
print("Maximum is:", dataframe_nfhs['Women who are literate'].max())
print("Mean is:", dataframe_nfhs['Women who are literate'].mean())
print("Standard Deviation is:", dataframe_nfhs['Women who are literate'].std())
print("Median is:", dataframe_nfhs['Women who are literate'].median())
```

```python
#Groupby

#Syntax
#df.groupby(by = []).agg(dict)

dictagg = {'All women age 15-49 years who are anaemic': 'mean'}
df_div_agg = dataframe_nfhs.groupby(by = ['Division']).agg(dictagg)
```

```python
dictagg = {'All women age 15-49 years who are anaemic': 'max', 'Households using clean fuel for cooking' : 'count'}
df_div_agg = dataframe_nfhs.groupby(by = ['Division']).agg(dictagg)
df_div_agg.head(5)
```

```python
# Join two dataframes

# Syntax
# pd.merge(left =, right =, left_on =, right_on =, how = '')

dataframe_nfhs_edit = dataframe_nfhs[['District', 'Division', 'All women age 15-49 years who are anaemic']]

df_pop = pd.read_excel('/content/Dist_Nfhs5_Sample.xlsx', sheet_name = 'Population')

df_merge = pd.merge(left = dataframe_nfhs_edit,
                    right = df_pop,
                    left_on = 'District',
                    right_on = 'District',
                    how = 'left')   #left,right,inner,outer(like pie charts)
df_merge
```
